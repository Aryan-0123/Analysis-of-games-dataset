```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
app_df=pd.read_csv('appstore_games.csv')
```

```python
app_df
```

```python
app_df.describe()
```

```python
#related info
app_df.info()
```

```python
app_df.drop_duplicates(subset=['Name'],inplace=True)
```

```python
app_df.drop(columns=['URL','Subtitle','Icon URL','In-app Purchases'],inplace=True)
app_df.reset_index(drop=True,inplace=True)
```

```python
app_df
```

```python
#removing null
pd.isnull(app_df).sum()
```

```python
app_df[['Average User Rating','User Rating Count','Price','Size']]=app_df[['Average User Rating','User Rating Count','Price','Size']].fillna(0.0)
```

```python
app_df['Languages'].fillna('Unkown',inplace=True)
```

```python
pd.isnull(app_df).sum()
```

```python
#data Visualization
sns.set_theme(style='darkgrid',palette='deep')
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (8,4)
matplotlib.rcParams['figure.facecolor'] = '#FFFFFF'
```

```python
pr=app_df.groupby(by='Price',as_index=False).agg(count=('Price','count')).head()
pr
```

```python
#plotting games vs price graph
plt.title("No of games for a given price")

v1=sns.barplot(x='Price',y='count',data=pr);
v1.set(xlabel='Price',ylabel='No_ofgames');
for p in v1.patches:
             v1.annotate("%.f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='black', xytext=(0, 20),
                 textcoords='offset points')
```

```python
#dividing games on the base of categories
app_df.Genres.replace('Games,','',regex=True,inplace=True)
app_df.loc[app_df.Genres.str.contains('Puzzle'),'Genres']='Strategy'
app_df.loc[app_df.Genres.str.contains('Board'),'Genres']='Strategy'
app_df.loc[app_df.Genres.str.contains('Action'),'Genres']='Action'
app_df.loc[app_df.Genres.str.contains('Adventure'),'Genres']='Adventure'
app_df.loc[app_df.Genres.str.contains('Family'),'Genres']='Family'
app_df.loc[app_df.Genres.str.contains('Education'),'Genres']='Family'
app_df.loc[app_df.Genres.str.contains('Casual'),'Genres']='Family'
app_df.loc[app_df.Genres.str.contains('Strategy'),'Genres']='Strategy'
```

```python
v1=sns.countplot(data=app_df,x='Genres');
plt.title('No of games in a Genre')
v1.set(xlabel="Genres",ylabel='No of games')
#To display the numbers
for p in v1.patches:
             v1.annotate("%.f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='black', xytext=(0, 20),
                 textcoords='offset points')
```

```python
#strategic games are most popular amongst the dataset
#plotting graph for avg ratings per genre
plt.xticks(rotation=65)
plt.title('Avg Rating in Each Genres')
sns.barplot(y='Average User Rating',x='Primary Genre',data=app_df);
```

```python
#plotting games on the base of number of ratings
age=sns.countplot(x='Average User Rating',data=app_df)
plt.title("No of Games based on Rating")
age.set(xlabel='Average User Rating',ylabel='No of games');
for p in age.patches:
             age.annotate("%.f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=11, color='black', xytext=(0, 20),
                 textcoords='offset points')
```

```python
#Age wise games with Genre
v1=sns.histplot(data=app_df,x='Age Rating',hue='Genres',multiple='dodge',shrink=0.8,palette='bright');
plt.title('Games Age wise according to Genre')
for p in v1.patches:
             v1.annotate("%.f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=9, color='black', xytext=(0, 20),
                 textcoords='offset points')
```

```python
#User rating of games according to Genre
v1=sns.scatterplot(data=app_df,x='Average User Rating',y='User Rating Count',hue='Genres',palette='deep');
plt.title("User Rating Count for Each Genres");
v1.set(ylim=(0,3032734))
```
